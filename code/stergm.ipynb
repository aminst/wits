```r
el_2011 = read.csv("../data/2011/edgelist_disparity.csv", stringsAsFactors = FALSE)
nl_2011 = read.csv("../data/2011/nodelist.csv", stringsAsFactors = FALSE)

nl_2011$gdp_us_dollar <- log(nl_2011$gdp_us_dollar)
nl_2011$area <- log(nl_2011$area)
nl_2011$population <- log(nl_2011$population)
nl_2011$gdp_per_capita <- log(nl_2011$gdp_per_capita)
nl_2011$gdp_us_dollar = as.numeric(scale(nl_2011$gdp_us_dollar))
nl_2011$gdp_growth = as.numeric(scale(nl_2011$gdp_growth))
nl_2011$inflation_rate = as.numeric(scale(nl_2011$inflation_rate))
nl_2011$population = as.numeric(scale(nl_2011$population))
nl_2011$gdp_per_capita = as.numeric(scale(nl_2011$gdp_per_capita))
nl_2011$agriculture_forestry_fishing_of_gdp = as.numeric(scale(nl_2011$agriculture_forestry_fishing_of_gdp))
nl_2011$industry_of_gdp = as.numeric(scale(nl_2011$industry_of_gdp))
nl_2011$merchandise_of_gdp = as.numeric(scale(nl_2011$merchandise_of_gdp))
nl_2011$net_barter_of_trade = as.numeric(scale(nl_2011$net_barter_of_trade))
nl_2011$foreign_direct_investment_inflows = as.numeric(scale(nl_2011$foreign_direct_investment_inflows))

net_2011 = network(el_2011, directed = TRUE, matrix.type = 'edgelist', vertex.attr=nl_2011, vertex.attrnames=colnames(nl_2011), ignore.eval = FALSE, names.eval='trade')
set.edge.value(net_2011, "trade", el_2011$weight)
net_2011
```

```r
stergm <- stergm(
                list(net_2011, net_2013, net_2015, net_2017),
                formation = ~ edges +
                        mutual +
                        triangles + 
                        nodecov("gdp_us_dollar") +
                        absdiff("gdp_us_dollar") +
                        nodecov("inflation_rate") +
                        absdiff("inflation_rate") +
                        nodecov("gdp_growth") +
                        absdiff("gdp_growth") +
                        nodematch("landlocked") +
                        nodematch("langoff_1") +
                        nodecov("agriculture_forestry_fishing_of_gdp") +
                        absdiff("agriculture_forestry_fishing_of_gdp") +
                        nodecov("industry_of_gdp") +
                        absdiff("industry_of_gdp") +
                        nodecov("merchandise_of_gdp") +
                        absdiff("merchandise_of_gdp") +
                        nodecov("net_barter_of_trade") +
                        absdiff("net_barter_of_trade") +
                        nodecov("foreign_direct_investment_inflows") +
                        absdiff("foreign_direct_investment_inflows"),
                dissolution = ~ edges +
                        mutual +
                        triangles + 
                        nodecov("gdp_us_dollar") +
                        absdiff("gdp_us_dollar") +
                        nodecov("inflation_rate") +
                        absdiff("inflation_rate") +
                        nodecov("gdp_growth") +
                        absdiff("gdp_growth") +
                        nodematch("landlocked") +
                        nodematch("langoff_1") +
                        nodecov("agriculture_forestry_fishing_of_gdp") +
                        absdiff("agriculture_forestry_fishing_of_gdp") +
                        nodecov("industry_of_gdp") +
                        absdiff("industry_of_gdp") +
                        nodecov("merchandise_of_gdp") +
                        absdiff("merchandise_of_gdp") +
                        nodecov("net_barter_of_trade") +
                        absdiff("net_barter_of_trade") +
                        nodecov("foreign_direct_investment_inflows") +
                        absdiff("foreign_direct_investment_inflows"),
                estimate="CMLE"
    )
```
